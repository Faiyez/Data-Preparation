```python
# Date Cleaning
import pandas as pd 
ndf = pd.read_csv("covid19_database.csv", index_col=['date']).sort_index()
ndf.head(50)
```

```python
#Working with total cases for canada and newfoundland
#alberta_values = ndf[(ndf['prname'] == 'Alberta')]
#bc_values = ndf[(ndf['prname'] == 'British Columbia')]
nl_values = ndf[(ndf['prname']) == 'Newfoundland and Labrador']
ca_values = ndf[(ndf['prname'] == 'Canada')]
n_df = ca_values
#n_df['alberta_totalCases'] = alberta_values['totalcases']
#n_df['bc_totalCases'] = bc_values['totalcases']
n_df['nl_totalCases'] = nl_values['totalcases']
n_df['ca_totalCases'] = ca_values['totalcases']
n_df['nl_rateDeathsLast7'] = nl_values['ratedeaths_last7']
n_df['ca_rateDeathsLast7'] = ca_values['ratedeaths_last7']
n_df = n_df[(n_df['reporting_year'] == 2022)]
n_df = n_df.drop(columns = ['pruid','prnameFR','prname','update','numtotal_last7','ratecases_total','ratecases_last7','ratedeaths_last7','numtotal_last14','numdeaths_last14','ratetotal_last14','ratedeaths_last14','avgcases_last7','avgincidence_last7','avgdeaths_last7','avgratedeaths_last7','numdeaths_last7','ratedeaths'])
column_names = ['nl_totalCases','nl_rateDeathsLast7','ca_totalCases','ca_rateDeathsLast7','reporting_week','reporting_year','totalcases','numdeaths']
n_df = n_df.reindex(columns=column_names)
n_df.head(100)
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2,1,figsize=(22,10))
axs = axs.flatten()
n_df['nl_rateDeathsLast7'].plot(ax=axs[0],title='NL Rate of Deaths Weekly in 2022',xlabel='Date')
n_df['ca_rateDeathsLast7'].plot(ax=axs[1],title='Canada Rate of Deaths Weekly in 2022',xlabel='Date')
```

```python
stats = ['mean','median','std','var']
agg_dict={'nl_rateDeathsLast7':stats,'ca_rateDeathsLast7':stats}
desc_stats = n_df.agg(agg_dict)
desc_stats
```
